```python
###My own equation
import math
import matplotlib.pyplot as plt
import numpy as np

N = 1.794
Lt_100 = 100
Eps_50 = 169.1
Eps_100 = 226.3
c_50 = -1189
c_100 = -764.9

kds = [ 0.1, 1, 2, 5, 10]


def fluo( kd, x, n, L, Eps, c):
    y = (kd + x/n*L + L -math.sqrt((kd + x/n*L + L)**2 -4*x*L/n*L))/2*Eps + c
    return y

xs = [0, 20, 30, 40, 50, 60 , 70, 80, 90, 100, 120, 140, 160, 180, 200, 300, 400, 500, 600, 700, 800, 900]
xrs = [x/100 for x in xs]

xrs = np.linspace(0,10,50)

yss = []
for kd in kds:
    ys = []
    for x in xrs:
        y = fluo(kd, x, N, Lt_100, Eps_100, c_100)
        ys.append(y)
    yss.append(ys)
    plt.plot(xrs, ys, label = str(kd))

workdir = '/mnt/e/'
with open(workdir + 'stimulated.txt', 'w') as f:
    f.write('\t'.join([str(x) for x in xrs]) + '\n')
    for ys in yss:
        f.write('\t'.join([str(y) for y in ys]) + '\n')

plt.legend()
plt.show()
    

```
